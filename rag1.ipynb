```python
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
docs_dir_path = "/Users/yuvraj/Downloads/llm_engineering-main/week5/knowledge-base"
vector_database = "/Users/yuvraj/Downloads/llm_engineering-main/week5/vector_db"
collection_name = "document_collection" 
```

```python
#loading the embedding model
embedding = HuggingFaceEmbeddings()
```

```python
# directory_loader
loader = DirectoryLoader(
    path = docs_dir_path,
    glob = "./*.pdf",
    loader_cls = UnstructuredFileLoader
)

# single file also can be loaded with UnstructuredFileLoader
```

```python
# load the documents
documents = loader.load()
```

```python
#initializin the text splitter
text_splitter = CharacterTextSplitter(
    chunk_size = 2000,
    chunk_overlap = 500
)
```

```python
# splittin the text into smaller chunk
text_chunks = text_splitter.split_documents(documents)
```

```python
len(text_chunks)
```

```python
# creating a vector space
vector_store = Chroma.from_documents(
    documents = text_chunks,
    embedding = embedding,
    persist_directory = vector_database,
    collection_name = collection_name
)
```
